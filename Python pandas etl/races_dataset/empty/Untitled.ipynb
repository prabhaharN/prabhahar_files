```python
import pandas as pd
```

```python
drivers_df=pd.read_parquet('drive____________________df')
races_df=pd.read_parquet('races____________________df')
circuits_df=pd.read_parquet('circuits____________________df')
constructors_df=pd.read_parquet('constructors____________________df')
pit_stops_df=pd.read_parquet('pit_stops____________________df')
results_df=pd.read_parquet('results____________________df')
qualifying_df=pd.read_parquet('qualifying____________________df')
lap_times_df=pd.read_parquet('qualifying____________________df')
```

```python
df=[drivers_df,races_df,constructors_df,circuits_df,results_df,qualifying_df,lap_times_df,pit_stops_df]
for i in df:
    i.drop(columns=['data_source','ingestion_date'],inplace=True)
```

```python
def merge(df1,df2,how,column):
    df=pd.merge(df1,df2,how=how,on=column)
    return df

def drop_column(df,column):
    df.drop(columns=column, inplace=True)
    return df
def rename_column(df,old_col_name, new_col_name):
    column_dict = dict(zip(old_col_name, new_col_name))
    df.rename(columns=column_dict, inplace=True)
    return df
```

```python
df_race_circuit=merge(races_df,circuits_df,'left','circuit_id')
df_race_circuit=drop_column(df_race_circuit,['circuit_id','circuit_ref','name_y','country','lattitude','longitude','altitude','round'])
df_race_circuit = rename_column(df_race_circuit,['location','year','name_x','date','time'],['circuit_location','race_year','race_name','race_date','race_time'])  
```

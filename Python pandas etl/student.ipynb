```python
import pandas as pd
```

```python
df=pd.read_excel('student.xlsx')
df
```

```python
full_name = df['name'].str.split(' ')
f=[]
for i in range(len(df)):
   f.append(full_name[i][0])
df['first_name']=f
```

```python

for i in range(len(df)):
    full_name[i].remove(full_name[i][0])
```

```python
d=[] 
for i in range(len(df)):
    if len(full_name[i])>1:
        d.append(full_name[i][0]+full_name[i][1])
    elif len(full_name[i])>0:
        d.append(full_name[i][0])
    else:
        d.append('')

df['last_name']=d

df
```

```python
df.drop('name', axis=1,inplace=True)
df
```

```python
d = df.pop('first_name')
```

```python
c = df.pop('last_name')
```

```python
df.insert(1,'first_name',d)
```

```python
df.insert(2,'last_name',c)
```

```python
grouped_count = df.groupby('class')['id'].count()
```

```python
grouped_sum = df.groupby('class')['mark'].sum()
```

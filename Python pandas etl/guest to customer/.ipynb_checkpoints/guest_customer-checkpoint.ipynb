```python
import pandas as pd
```

```python
pf_guest=pd.read_excel('guest.xlsx')
pf_guest
```

```python
pf_cust=pd.read_excel('customer.xlsx')
cust_rn=pf_cust.rename(columns = {'c_name':'name','c_city':'city','c_phone':'phone'})
cust_rn
```

```python
df_call_plan = pd.DataFrame({ 
    'Guest_name': [], 
    'Guest_city': [], 
    'Guest_phone': []
})
```

```python
df_call_plan
```

```python
all_rows = pf_guest.merge(cust_rn.drop_duplicates(), on=[
    'name','city','phone'], how='left', indicator=True)

rows_not_in = all_rows[all_rows['_merge'] == 'left_only']
rows_not_in = rows_not_in.drop(['_merge'], axis=1)
rows_not_in=rows_not_in.drop(['pro_flag','c_id'],axis=1)

rows_not_in
```

```python
guest_rn=rows_not_in.rename(columns = {'name':'Guest_name','city':'Guest_city','phone':'Guest_phone'})
guest_rn
```

```python
df_call_plan = pd.concat([df_call_plan,guest_rn], ignore_index=True)
df_call_plan['Call_plan_id']= range(100,100+len(df_call_plan))
df_call_plan    
```
